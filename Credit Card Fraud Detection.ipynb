# Importing all the necessary packages and Data Set

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load the dataset
train_df = pd.read_csv('fraudTrain.csv')
test_df = pd.read_csv('fraudTest.csv')

# Inspect the data
train_df.head()
```

# Pre - processing the Dataset

### Train Data set

```python
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
train_df['dob'] = pd.to_datetime(train_df['dob'])
train_df['transaction_hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['transaction_day'] = train_df['trans_date_trans_time'].dt.day
train_df['transaction_month'] = train_df['trans_date_trans_time'].dt.month
train_df['transaction_dayofweek'] = train_df['trans_date_trans_time'].dt.dayofweek
train_df['age'] = (pd.to_datetime('today') - train_df['dob']).dt.days // 365
train_df = train_df.drop(['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],axis=1)
train_df
```

### Test Data set

```python
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])
test_df['dob'] = pd.to_datetime(test_df['dob'])
test_df['transaction_hour'] = test_df['trans_date_trans_time'].dt.hour
test_df['transaction_day'] = test_df['trans_date_trans_time'].dt.day
test_df['transaction_month'] = test_df['trans_date_trans_time'].dt.month
test_df['transaction_dayofweek'] = test_df['trans_date_trans_time'].dt.dayofweek
test_df['age'] = (pd.to_datetime('today') - test_df['dob']).dt.days // 365
test_df = test_df.drop(['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],axis=1)
test_df
```

```python
train_df = train_df.dropna().reset_index(drop=True)
train_df
```

```python
test_df = test_df.dropna().reset_index(drop=True)
test_df
```

# Transforming the values using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
columns_to_encode = ['merchant', 'category', 'gender', 'job']

for column in columns_to_encode:
    train_df[column] = encoder.fit_transform(train_df[column])
    test_df[column] = encoder.fit_transform(test_df[column])
```

```python
X_train = train_df.drop(['is_fraud'], axis=1)
y_train = train_df['is_fraud']
X_test= test_df.drop(['is_fraud'], axis=1)
y_test = test_df['is_fraud']
```

```python
train_df.head()
```

# Modeling

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
class_rep = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Classification Report :\n {class_rep}")
print("="*50)
```
